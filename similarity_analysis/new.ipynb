```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
from scipy.stats import wilcoxon, ttest_rel
from matplotlib.patches import Rectangle
# ---- aesthetics ----
stimuli = ["Familiar Stimuli", "Unfamiliar Stimuli", "Scrambled Stimuli"]
tasks   = ["Face Recognition", "Dual Task", "Object Recognition", "Untrained"]
markers = ['*', 'o', '^', 'd', 's', 'p', 'P', 'X']

original_colors = [
    ["#0000ff", "#3498db", "#26F7FD"],  # blue
    ["#ED0DD9", "#FF69B4", "#FFC0CB"],  # pink
    ["#009900", "#32CD32", "#66FF66"],  # green
    ["#808080", "#A9A9A9", "#CCCCCC"]   # gray
]
bar_colors = [
    original_colors[0],  # Face Recognition (blue)
    original_colors[2],  # Dual Task (green)
    original_colors[1],  # Object Recognition (pink)
    original_colors[3],  # Untrained (gray)
]

# ---- compute Wilcoxon stats ----
n_tasks, n_models, n_stim = region_data.shape
conds = [(t, s) for t in range(n_tasks) for s in range(n_stim)]
n_conds = len(conds)
stat_mat = np.zeros((n_conds, n_conds))
p_mat    = np.ones((n_conds, n_conds))

for i, (t1, s1) in enumerate(conds):
    for j, (t2, s2) in enumerate(conds):
        if i > j:
            x = region_data[t1, :, s1]
            y = region_data[t2, :, s2]
            w, p = ttest_rel(x, y) #mode="exact")
            stat_mat[i, j] = w
            p_mat[i, j] = p

from statsmodels.stats.multitest import multipletests
import numpy as np
from matplotlib.patches import Rectangle

# flatten lowertriangle p-values
idxs = np.tril_indices(n_conds, k=-1)
raw_ps = p_mat[idxs]

# choose a method: "holm" or "fdr_bh"
rejected, p_adj, _, _ = multipletests(raw_ps, alpha=0.01, method="fdr_bh")

# build an adjustedp matrix
p_mat_adj = np.ones_like(p_mat)
p_mat_adj[idxs] = p_adj


# ---- plot combined figure ----
fig = plt.figure(figsize=(18, 6))          # a bit wider overall
gs  = fig.add_gridspec(1, 2,               # 2 panels
                       width_ratios=[3, 2],   # <-- more space for heat-map
                       wspace=0.1)

# Panel A: grouped bar plot
ax_bar = fig.add_subplot(gs[0, 0])
group_centers = np.arange(n_tasks)
bar_width = 0.3
jitters= [np.random.uniform(-bar_width/3, bar_width/3) for _ in range(7)]
for ti in range(n_tasks):
    for si in range(n_stim):
        x = group_centers[ti] + (si - (n_stim-1)/2) * bar_width
        data = region_data[ti, :, si]
        mean = data.mean()
        sem  = data.std(ddof=1) / np.sqrt(n_models)
        ax_bar.bar(
            x, mean,
            yerr=sem,
            width=bar_width,
            color=bar_colors[ti][si],
            capsize=4
        )

        # overlay model points with a tiny jitter
        for mi, y in enumerate(data):
            ax_bar.scatter(
                x + jitters[mi], y,
                marker=markers[mi],
                edgecolor='None',
                color="black",
                alpha=0.6,
                s=50
            )

ax_bar.set_xticks(group_centers)
ax_bar.set_xticklabels(tasks)
ax_bar.tick_params(axis='both', length=0)
ax_bar.set_ylabel("Similarity Scores", fontsize=24)
ax_bar.set_ylim(0.2, 0.7)
# create custom legend for stimuli using dummy colored patches
handles = [Rectangle((0, 0), 1, 1, color=bar_colors[0][si]) for si in range(n_stim)]
ax_bar.legend(handles=handles, labels=stimuli, title="Stimulus", loc="lower left")

# Panel B: heatmap of t-statistic with stars for significance
ax_heat = fig.add_subplot(gs[0, 1])

# Remove first row and last column from the matrices
stat_mat_mod   = stat_mat[1:, :-1]
p_mat_adj_mod  = p_mat_adj[1:, :-1]
new_n = stat_mat_mod.shape[0]
im = ax_heat.imshow(stat_mat_mod, cmap="bwr", vmin=-11, vmax=11)
cbar = plt.colorbar(im, ax=ax_heat, fraction=0.046, pad=0.03)
cbar.set_ticks([10, 5, 0, -5, -10])
cbar.ax.tick_params(axis="y", which="both", length=0)
cbar.set_label("t-values")

# Make the unused upper-right part white (adjust indices back to original positions)
for i in range(new_n):
    for j in range(new_n):
        # Original row index is i+1; use that for the condition relative to original columns
        if (i + 1) <= j:
            ax_heat.add_patch(
                Rectangle((j - 0.5, i - 0.5), 1, 1,
                          facecolor="white", edgecolor="none", zorder=3)
            )

# Update tick labels.
# For rows, use original labels from the removed first row: indices 1 to end.
# For columns, use original labels from the kept columns: indices 0 to n_conds-2.
all_labels = [f"{tasks[t]}\n{stimuli[s]}" for t, s in conds]
row_labels = all_labels[1:]
col_labels = all_labels[:-1]

ax_heat.set_xticks(np.arange(new_n))
ax_heat.set_yticks(np.arange(new_n))
ax_heat.set_xticklabels(col_labels, rotation=90, fontsize=7)
ax_heat.set_yticklabels(row_labels, fontsize=7)
ax_heat.set_title("Pairwise Model-condition Tests")
ax_heat.tick_params(axis='both', length=0)
ax_heat.grid(False)

# Annotate lower triangle: if (original row index) > (original col index)
for i in range(new_n):
    for j in range(new_n):
        if (i + 1) > j:
            # Map back: row i corresponds to original row i+1; column j remains the same.
            p_corr = p_mat_adj[i+1, j]
            stat   = stat_mat[i+1, j]
            if   p_corr < 0.001:
                star = "***"
            elif p_corr < 0.01:
                star = "**"
            elif p_corr < 0.05:
                star = "*"
            else:
                star = ""
            txt = f"{star}\n{stat:.1f}" if star else f"{stat:.1f}"
            ax_heat.text(j, i, txt,
                         ha='center', va='center',
                         color="black", fontsize=8)

fig.suptitle("Fusiform - rh Region", fontsize=15)
plt.tight_layout()
plt.show()
# save the figure
fig.savefig(os.path.join("fusiform_rh_model_comparison.pdf"), format="pdf", bbox_inches='tight', dpi=600)
fig.savefig(os.path.join("fusiform_rh_model_comparison.svg"), format="svg", bbox_inches='tight', dpi=600)
```
